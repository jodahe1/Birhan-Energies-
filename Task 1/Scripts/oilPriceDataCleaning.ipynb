```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
# Read the CSV file with the appropriate encoding
data = pd.read_csv("../Data/oil-production-statistics.csv", encoding='latin-1')
print(data.head())
```

```python
# Remove rows with missing values
data = data.dropna()

# Remove any duplicate rows
data = data.drop_duplicates()
```

```python
# Convert 'year' column to integer data type
data['year'] = data['year'].astype(int)

# Clean up any whitespace in column names
data.columns = data.columns.str.strip()
```

```python

# Save the cleaned data to a new CSV file
data.to_csv('oil-production-statistics-cleaned_data.csv', index=False)
```

EDA

```python
# 1. Distribution of Oil Production Values
plt.figure(figsize=(10, 6))
sns.histplot(data['value'], kde=True)
plt.title('Distribution of Oil Production Values')
plt.xlabel('Oil Production Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Oil Production by Country
plt.figure(figsize=(12, 6))
sns.barplot(x='country_name', y='value', data=data)
plt.title('Oil Production by Country')
plt.xlabel('Country')
plt.ylabel('Oil Production Value')
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Oil Production by Product
plt.figure(figsize=(10, 6))
sns.boxplot(x='product', y='value', data=data)
plt.title('Oil Production by Product')
plt.xlabel('Product')
plt.ylabel('Oil Production Value')
plt.show()
```

```python
# 4. Oil Production Trend Over Time
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='value', data=data, hue='country_name')
plt.title('Oil Production Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Oil Production Value')
plt.show()
```

```python
# . Pair Plot
plt.figure(figsize=(10, 6))
sns.pairplot(data, hue='product')
plt.title('Pair Plot of Oil Production Data')
plt.show()
```
